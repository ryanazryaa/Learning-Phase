```python
#EXERCISE 9
import random

the_number = random.randint(1,9)
guess_number = int(input("Guess a number between 1 and 9: "))
exit = False

while exit == False:
  if guess_number == the_number:
    print("You guessed the number")
    again = str(input("Do you want to play again? Y/N: ")).lower()
    if again == "y":
      exit = False
      the_number = random.randint(1,9)
      guess_number = int(input("Guess a number between 1 and 9: "))
    elif again == "n":
      break
  elif guess_number > the_number:
    print("Too high")
    guess_number = int(input("Guess a number between 1 and 9: "))
  elif guess_number < the_number:
    print("Too low")
    guess_number = int(input("Guess a number between 1 and 9: "))

```
