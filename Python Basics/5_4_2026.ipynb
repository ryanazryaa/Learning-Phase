```python
#EXERCISE 11
number = int(input("Enter a number: "))
divisor_count = 0

for i in range(1, number + 1):
  if number % i == 0:
    divisor_count += 1
if divisor_count == 2 or number == 1:
  print("Your number is prime")
else:
  print("Your number is not prime")
```

```python
#EXERCISE 13

def fibonacci(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter a number: "))

if n < 0:
  print("Please enter a non-negative number.")
else:
  print(f"The first {n} Fibonacci numbers are:")
  for i in range(n):
    print(fibonacci(i), end=" ")
  print()
```

```python
#EXERCISE 14
import random

def random_list():
  list_numbers = list(random.sample(range(30), random.randint(2, 10)))
  return list_numbers

def no_duplicates(list1):
  new_list = []
  for i in list1:
    if i not in new_list:
      new_list.append(i)
  return new_list

list1 = random_list()
list2 = no_duplicates(list1)
print(list2)
```

```python
#EXERCISE 14 EXTRAS 1
import random

def random_list():
  list_numbers = list(random.sample(range(30), random.randint(2, 10)))
  return list_numbers
def no_dupes(list1):
  return list(set(list1))

list1 = random_list()
list2 = no_dupes(list1)
print(list2)
```

```python
#EXERCISE 15

sentence = str(input("Enter a sentence: "))
def split(sentence):
  return list(sentence.split())

def reverse(new_sentence):
  return new_sentence[::-1]

new_sentence = split(sentence)
reverse_sentence = reverse(new_sentence)
print(*reverse_sentence)
```

```python
#EXERCISE 16 EXTRA 1
import random

s = "abcdefghijklmnopqrstuvwxyz01234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()?"
weak = ["apple", "banana", "cherry", "pineapple", "orange"]
pass_strength = str(input("Is the password strong or weak? ")).lower()

if pass_strength == "strong":
  def passlength():
    return int(input("How long would you want your password? "))
  def passmaker(passlen):
    return "".join(random.sample(s, passlen))
  passlen = passlength()
  password = passmaker(passlen)
  print(password)
elif pass_strength == "weak":
  def passlength():
    return int(input("How long would you want your password? "))

  def weakpass1():
    return random.choice(weak)

  def weakpass2():
    return "".join(random.sample(weak, 2))

  passlen = passlength()

  if passlen == 2:
    password = weakpass2()

    print(password)
  elif passlen == 1:
    password = weakpass1()
    print(password)
  else:
    print("For a weak password, please choose length 1 or 2.")
```
